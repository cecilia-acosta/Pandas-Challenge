```python
import pandas as pd
```

## Purchasing Analysis (Total)

```python
Unique_Items=purchase_df["Item ID"].nunique()
Unique_Items
```

```python
Average_Price=purchase_df["Price"].mean()
Average_Price
```

```python
Number_of_Purchases=purchase_df["Purchase ID"].count()
Number_of_Purchases
```

```python
Total_Revenue=purchase_df["Price"].sum()
Total_Revenue
```

```python
Purchasing_Analysis_Total=pd.DataFrame([{
    "Number of Unique Items":Unique_Items,
    "Average Price":Average_Price,
    "Number of Purchases":Number_of_Purchases,
    "Total Revenue":Total_Revenue
}])
Purchasing_Analysis_Total
```

```python
Purchasing_Analysis_Total["Average Price"]=Purchasing_Analysis_Total["Average Price"].map("${:.2f}".format)
```

```python
Purchasing_Analysis_Total["Total Revenue"]=Purchasing_Analysis_Total["Total Revenue"].map("${:,.2f}".format)
```

```python
Purchasing_Analysis_Total
```

##  Gender Demographics

```python
purchase_df.head()
```

```python
Gender_df=purchase_df[["SN","Gender"]]
Gender_df.head()
```

```python
Total_Gender=Gender_df["SN"].nunique()
Total_Gender
```

```python
Gender_clean=Gender_df.drop_duplicates(["SN"],keep="first",inplace=False)
Gender_clean
```

```python
#Gender_du_group=Gender_du.groupby(["Gender"])

Gender_demographics=pd.DataFrame({
    "Total Count":Gender_clean.groupby("Gender")["SN"].count(),
    "Percentage of Players":Gender_clean.groupby("Gender")["SN"].count()/Total_Gender
})

Gender_demographics
```

```python
Gender_demographics["Percentage of Players"]=Gender_demographics["Percentage of Players"].map("{:.2%}".format)
```

```python
Gender_demographics.sort_values("Total Count",ascending=False)
```

## Age Demographics

```python
purchase_df.head()
```

```python
Age_df=purchase_df[["SN","Age"]]
Age_df.head()
```

```python
Age_clean=Age_df.drop_duplicates(["SN"],keep="first",inplace=False)
Age_clean
```

```python
Age_Total=Age_clean["SN"].nunique()
Age_Total
```

```python
bins = [0,9,14,19,24,29,34,39,100]
age_buckets = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
Age_clean["Age Buckets"]=pd.cut(Age_clean["Age"],bins,labels=age_buckets)
Age_clean
                                    
```

```python
Age_demographics=pd.DataFrame({
    "Total Count":Age_clean.groupby("Age Buckets")["SN"].count(),
    "Percentage of Players":Age_clean.groupby("Age Buckets")["SN"].count()/Age_Total
})

Age_demographics
```

```python
Age_demographics["Percentage of Players"]=Age_demographics["Percentage of Players"].map("{:,.2%}".format)
```

```python
Age_demographics
```

##  Purchasing Analysis (Age)

```python
purchase_df.head()
```

```python
Age_df2=purchase_df[["Purchase ID","Age","Price"]]
Age_df2.head()
```

```python
bins = [0,9,14,19,24,29,34,39,100]
age_buckets = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
Age_df2["Age Buckets"]=pd.cut(Age_df2["Age"],bins,labels=age_buckets)
Age_df2
```

```python
Age_analysis=pd.DataFrame({
    "Purchase Count":Age_df2.groupby("Age Buckets")["Purchase ID"].count(),
    "Avg Purchase Price":Age_df2.groupby("Age Buckets")["Price"].mean(),
    "Total Purchase Value":Age_df2.groupby("Age Buckets")["Price"].sum(),    
    "Avg Purchase per Person":Age_df2.groupby("Age Buckets")["Price"].sum()/Age_demographics["Total Count"]
})
Age_analysis
```

```python
Age_analysis["Avg Purchase Price"]=Age_analysis["Avg Purchase Price"].map("${:.2f}".format)
Age_analysis["Total Purchase Value"]=Age_analysis["Total Purchase Value"].map("${:,.2f}".format)
Age_analysis["Avg Purchase per Person"]=Age_analysis["Avg Purchase per Person"].map("${:.2f}".format)
```

```python
Age_analysis
```

## Top Spenders

## Most Popular Items

```python
purchase_df.head()
```

```python
items_df=purchase_df[["Item Name","Item ID","Price"]]
items_df
```

```python
popular_items=pd.DataFrame({
    "Purchase Count":items_df.groupby(["Item ID","Item Name"])["Item ID"].count(),
    "Item Price":items_df.groupby(["Item ID","Item Name"])["Price"].mean(),
    "Total Purchase Value":items_df.groupby(["Item ID","Item Name"])["Price"].sum() 
})
popular_items
```
